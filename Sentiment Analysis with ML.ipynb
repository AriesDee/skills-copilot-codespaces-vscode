```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/anshupandey/Working_with_Large_Language_models/main/sentimentdata.csv"

df = pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
# accessing documents and lables
x = df['document']
y = df['label']
```

### Vectorization
Using TFIDF to preform vectorization

```python
### Vectorization
from sklearn.feature_extraction.text import TfidfVectorizer

vec = TfidfVectorizer(lowercase=True,stop_words='english')
vec.fit(x)
```

```python
print(len(vec.get_feature_names_out()))
```

```python
print(vec.get_feature_names_out())
```

```python
#Generate vector
xvec = vec.transform(x)
print(xvec.shape)
```

```python
for i in range(len(x)):
  print(x[i])
  print(xvec[i:])
```

### Sentiment Analysis with ML

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()

# train the model with vectorized data and labbes
model.fit(xvec,y)
```

```python
ip = "I love pasta"
ipvec = vec.transform([ip])
print(ipvec)
```

```python
model.predict(ipvec)
```

```python
ip = "I hate pasta"

ipvec = vec.transform([ip]).toarray()
model.predict(ipvec)
```
